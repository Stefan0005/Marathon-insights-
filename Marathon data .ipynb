```python
#Race speed vs seasons how the weather seaons affect athelete speed using lamda function to map the weather seasons 
#Spring 3-5
#Summer 6-8
#Fall 9-11
#winter 12-2
#rounded up to 2 decimal places 
```

```python
df2['race_month']=df2['race day'].str.split('.').str.get(1).astype(int)
```

```python
df2['race_season']=df2['race_month'].apply(lambda x: 'winter'if x > 11 else 'fall'if x > 8 else 'summer'if x > 5 else 'spring' if x >2 else 'winter')
```

```python
df2.tail(30)
```

```python
#Average speed and number of runners by seasons to give a consolidated view on how the weather seasons affect them as a whole 
df2.groupby('race_season')['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending=False)
```
